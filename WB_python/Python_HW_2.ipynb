## Рассмотрим сортировку вставкой

```python
numbers = [random.randint(0, 10000) for i in range(20000)]
numbers
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >=0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr



pure_python_insertion = insertion_sort(numbers)
print(pure_python_insertion, '\n')

```

```python
%timeit insertion_sort(numbers)
```

```python
def insertion_sort_numpy(arr):
    arr = np.array(arr)
    for i in range(1, arr.shape[0]):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

my_np_insertion = insertion_sort_numpy(numbers)
print(my_np_insertion, '\n',)
%timeit insertion_sort_numpy(numbers)
```

```python
np_insertion = np.sort(numbers) 
print(np_insertion)
%timeit np.sort(numbers)
```

<h1>Задание 3</h1> 

У нас есть данные о потреблении алкоголя о потреблнеии Алкоголя в мире https://github.com/fivethirtyeight/data/tree/master/alcohol-consumption. 

Давайте представим эти данные в виде структурированных массивов и определим, где же люди пьют больше вина, пива и других спиртных напитков.

```python
import numpy as np

with open('drinks.csv', 'r') as f:
    data = f.readlines()[1:]
    
countries = []
beer_servings = []
spirit_servings = []
wine_servings = []
total_litres_of_pure_alcohol = []

for line in data:
    country, beer, spirit, wine, total = line.strip().split(',')
    countries.append(country)
    beer_servings.append(int(beer))
    spirit_servings.append(int(spirit))
    wine_servings.append(int(wine))
    total_litres_of_pure_alcohol.append(float(total))

countries = np.array(countries)
beer_servings = np.array(beer_servings)
spirit_servings = np.array(spirit_servings)
wine_servings = np.array(wine_servings)
total_litres_of_pure_alcohol = np.array(total_litres_of_pure_alcohol)
```

```python
max_beer_idx = np.argmax(beer_servings)
max_beer_country = countries[max_beer_idx]

max_spirit_idx = np.argmax(spirit_servings)
max_spirit_country = countries[max_spirit_idx]

max_wine_idx = np.argmax(wine_servings)
max_wine_country = countries[max_wine_idx]
```

```python
print(f"Пьют больше всего пива: {max_beer_country}")
print(f"Пьют больше всего спиртных напитков: {max_spirit_country}")
print(f"Пьют больше всего вина: {max_wine_country}")
```

<h1>Задание 4</h1> 

У нас есть Полный каталог всех случаев, когда кто-то проклинал или истекал кровью в фильме Квентина Тарантино. https://github.com/fivethirtyeight/data/tree/master/tarantino

Давайте представим эти данные в виде структурированных массивов и определим:

- число проклятий, летальных исходов и их отношение по фильмам 

- частота употребления конкретных прокрятий 

- распределение времени между проклятиями 

```python
import numpy as np


data = np.genfromtxt('tarantino.csv', delimiter=',', dtype=None, names=True, encoding='utf-8')


sorted_data = np.sort(data, order='minutes_in') 


unique_movies = np.unique(sorted_data['movie'])
for movie in unique_movies:
    movie_data = sorted_data[sorted_data['movie'] == movie]
    curses_count = np.sum(movie_data['type'] == 'word')
    deaths_count = np.sum(movie_data['type'] == 'death')
    print(f"Movie: {movie}")
    print(f"Curses: {curses_count}, Deaths: {deaths_count}, Ratio: {curses_count/deaths_count if deaths_count != 0 else 0}")


unique_words = np.unique(sorted_data['word'])
for word in unique_words:
    word_count = np.sum(sorted_data['word'] == word)
    print(f"Word: {word}, Count: {word_count}")


times = sorted_data['minutes_in']
time_diffs = np.diff(times)
print("Average time difference between curses:", np.mean(time_diffs))
print("Minimum time difference between curses:", np.min(time_diffs))
print("Maximum time difference between curses:", np.max(time_diffs))
```
